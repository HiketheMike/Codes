# Riskfolio introduction

```python
import riskfolio as rper
pd.options.display.float_format = '{:.5f}'.format
```

```python
start = '2016-01-01'
end = '2019-12-30'

assets = ['AG', 'TG', 'CPB', 'MO', 'APA','JPM', 'PSA', 
          'BAX', 'BMY', 'LUV', 'PCAR', 'TXT', 'TWO', 'DE', 'MSFT', 'HPQ', 'SEE', 
          'VZ', 'CNP', 'NI', 'T', 'BA']
assets.sort()

# Download the data
df = yf.download(assets, start, end, auto_adjust = False)
data = df['Adj Close'] 
data.columns = assets
```

```python
# Calculate Returns
Y = data[assets].pct_change().dropna()
Y
```

```python
# Estimate Mean Variance Portfolios
import riskfolio as rp
port = repr.Portfolio(returns = Y)

# Calculate the optimal Portfolio
method_mu = 'hist'
method_cov = 'hist'
port.assets_stats(method_mu = method_mu, method_cov = method_cov)
model = 'Classic'
rm = 'MV'
obj = 'Sharpe'
hist = True
rf = 0
l = 0

w = port.optimization(model = model, rm = rm, obj = obj, rf = rf, l = l, hist = hist)
display(w.T)
```

```python
# Plot the Weights as a Donut Chart
ax = rp.plot_pie(w = w, title = 'Sharpe Mean-Variance', cmap = 'tab20', height = 6, width = 10, ax = None)
```

```python
points = 500
frontier = port.efficient_frontier(model = model, rm = rm, points = points, hist = hist)
label = 'Max Risk Adjusted Return Portfolio'
mu = port.mu
cov = port.cov
returns = port.returns

ax = rp.plot_frontier(w_frontier = frontier, mu = mu, cov = cov, returns = returns, alpha = 0.05, w = w, label = label, marker = '*', s = 16, c = 'r', height = 6, width = 10, ax = None)
```

```python
ax = rp.plot_frontier_area(w_frontier = frontier, cmap = 'tab20', height = 6, width = 10, ax = None)
```
