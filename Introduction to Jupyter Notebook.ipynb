```python
# Let's start by importing some essential libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Code cell example:
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot the sine curve
plt.plot(x, y)
plt.title(" Sine Wave ")
plt.xlabel("x")
plt.ylabel("$sin(x)$")
plt.show()
```

```python
# Section 3: Tips for Efficiency in Jupyter
# These are a set of shortcuts that can help streamline your work
```

 - Use **Shift + Enter** to run the current cell and move to the next.
 - Use **Ctrl + Enter** to run the current cell and stay on the same cell.
 - Press **Esc** to exit edit mode, and then use commands like 'A' to insert a cell above or 'B' to insert below.
